```python
import pandas as pd;
```

```python
data=pd.read_csv("/content/drive/MyDrive/dnn_data.csv")
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

le.fit(data['label'])
data['label']=le.transform(data['label'])
```

```python
data.head()
```

```python
feature_set=data[['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']]
#the target variable
target_var= data['label']
feature_set.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_set,target_var,test_size = 0.2,random_state =4)
X_train.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train,test_size = 0.2,random_state =4)
```

```python

from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=7, activation='relu'))
model.add(Dense(7, activation='relu'))


model.add(Dense(1, activation='sigmoid'))
```

```python
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# fit the keras model on the dataset
model_history=model.fit(X_test, y_test, epochs=10000, batch_size=2,validation_data=(X_val,y_val))
print(model_history)
```

```python
values_predicted = model.predict(X_test)
```

```python
score = model.evaluate(X_test, y_test,verbose=1)

print(score)
```
